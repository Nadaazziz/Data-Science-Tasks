```python
import pandas as pd

df = pd.read_csv('C:\\Users\\pc\\Downloads\\1) iris.csv', encoding='latin1')

df.head()
```

```python
print(df.describe())
print(df.isnull().sum())
```

```python
print( df.mean(numeric_only=True))
print( df.median(numeric_only=True))
print( df.var(numeric_only=True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df.hist(figsize=(10, 8), bins=15, color='skyblue')
plt.suptitle("Histograms of Numeric Features", fontsize=14)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.title("Boxplot of Numeric Features")
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=df, palette='viridis')
plt.title("Petal Length vs Petal Width by Species")
plt.show()
```

```python
sns.pairplot(df, hue='species', diag_kind='hist')
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```
